Creating Array

```python
import numpy as np
```

```python
print(np.zeros((2,2),dtype=int))    # 2x2 matrix storing only integers
print()
print(np.ones((3,3)))               # 3x3 matrix with all values as 1
print()          
print(np.full((3,3),10))            # fill the given number using full function
print()          
print(np.eye(3))                    # identity matrix 
print()          
print(np.arange(1,100))             # aranging numbers
print()          
print(np.linspace(0,1,5))           # 5 numbers between 0 and 1
print()
print(np.empty((5,5)))              # creates a empty spaces or any values in the matrix
```

Array attributes

```python
arr = np.array([[1 ,2 ,3],[4 ,5 ,6]])

print("shape of array\t:",arr.shape)  # shape of array

print("dimension\t: ",arr.ndim)   # dimension

print("Total elements\t: ",arr.size)   # Total elements

print("data type\t: ",arr.dtype)  # data type
```

mathematical functions

```python
arr1 = np.array([1 ,2 ,3])
arr2 = np.array([4 ,5 ,6])


print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

Statistical functions

```python
arr = np.array([1 , 2 ,3 ,4 ,5])

print(arr.mean())
print(arr.max())
print(arr.min())
print(arr.sum())
```
