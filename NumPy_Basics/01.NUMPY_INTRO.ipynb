Creating Array

```python
import numpy as np
```

```python
print(np.zeros((2,2),dtype='int64'))          # 2x2 matrix storing only integers
```

```python
print(np.ones((3,3)))                         # 3x3 matrix with all values as 1
```

```python
print(np.zeros((3,3)))                        # 3x3 matrix with all values as 1  
```

```python
print(np.zeros_like(np.ones((3,3))))          # 3x3 matrix with all values as 0 using zeros_like function     
```

```python
print(np.full((3,3),10))                      # fill the given number using full function     
```

```python
print(np.full_like(np.zeros((3,3)),10))       # fill the given number using full function with the help of zeros_like function
```

```python
print(np.eye(3))                              # identity matrix 
```

```python
print(np.identity(5))                       # identity matrix with the help of identity function
```

```python
print(np.arange(1,100))                       # aranging numbers        
```

```python
print(np.linspace(0,1,5))                     # 5 numbers between 0 and 1
```

```python
print(np.empty((5,5)))                        # creates a empty spaces or any values in the matrix
```

```python
arr = np.array([[1 ,2 ,3]])                   # creating a numpy array
repeat = np.repeat(arr,2 , axis = 0)          # repeat the given number for 2 times
print(repeat)
```
