rand - genarate random float between 0 and 1
randint - Generate a random integer


```python
import numpy as np

# np.random.seed(10)  # reprduce same input every time the program excutes 
print(np.random.rand(5))
print(np.random.randint(50,100))
print(np.random.randint(50,100, size = 5))  # size = 5 - creates a list of 5 numbers [1 ,2 ,3 ,4 ,5]
```

RANDOM DISTRIBUTIONS - machine learning

1.normal distributions - loc= - average --> parameter
                         scale= - std deviation
                         size= - no.datapoints

2.Uniform distributions - ( lowvalue , highvalue , size )
3.binomial distributions

```python
import numpy as np
print("normal: ",np.random.normal(loc= 50, scale= 15, size= 10))
print("unifrom: ",np.random.uniform(low=5,high=15,size=5))

```
