```python
# scalar arithmetic

import numpy as np
```

```python
arr = np.array([1 ,2 ,3])

print(arr + 1)
print(arr - 2)
print(arr * 3)
print(arr / 4)
print(arr ** 5)
```

```python
# Vectorised math functions

arr = np.array([1.01 ,2.5 ,3.99])

print(np.sqrt(arr))
print()

print(np.round(arr))
print()

print(np.ceil(arr))
print()

print(np.floor(arr))
print()

print("pi value: ",np.pi)
print()
```

```python
# arthmetic operators

arr1 = np.array([1 ,2 ,3])
arr2 = np.array([4 ,5 ,6])

print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
print(arr1 ** arr2)
print(arr1 // arr2)
print(arr1 % arr2)
```

```python
import numpy as np

# comparison operators

scores = np.array([91 ,55 ,100 ,73 ,82 ,64])

print(scores == 100)
print(scores >= 60)
print(scores <  60)


scores[scores < 60] = 0
print(scores)
```
