Matrix Multiplication

```python
import numpy as np 

# multiplication

A = np.array([[1 ,2 ,3],[4 ,5 ,6]]) 
B = np.array([[7 ,8],[9 ,10 ],[11 ,12]]) 

C = A @ B  # @ = muliplication
print("C: ",C)

# another method

X = np.matmul(A,B) # matmul - multiplication

print("X: ",X)
  
# another method

Y = np.dot(A,B) # dot - multiplication

print("Y: ",Y)  
```

inverse of matrix

```python
import numpy as np
from numpy.linalg import inv

A = np.array([[4,7],[2,6]])

inverse_A = inv(A)
print(inverse_A)
```

determinant

```python
import numpy as np
from numpy.linalg import det
A = np.array([[4,6],[7,6]])
deter = det(A)
print(round(deter,3))
```
