WEATHER DATA ANALYSIS

~ Analyse temparature data for a month using Numpy arrays


```python
import numpy as np

np.random.seed(0)
temp = np.random.randint(low=25,high=35,size=(30))

print("temparature by days: ",temp)
print("avarage of temp\t: ",np.mean(temp))
print("Median Temp\t: ",np.median(temp))
print("Max temp\t: ",np.max(temp))
print("Min temp\t: ",np.min(temp))

diff = np.diff(temp)
print("daily temp difference\t: ",diff)

hottest_day = np.argmax(temp) + 1
print("hottest_day\t: ",hottest_day)

coldest_day = np.argmin(temp) + 1
print("coldest_day\t: ",coldest_day)
```
