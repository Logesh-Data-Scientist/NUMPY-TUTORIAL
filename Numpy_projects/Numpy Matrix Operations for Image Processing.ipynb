Numpy Matrix Operations for Image Processing Stimulation

~ stimulate an image (matrix) and apply brightness and contrast operations

```python
import numpy as np

# creates a 5 x 5 matrix grayscale image (pixel values 0-255)

image = np.random.randint(0,256,(5,5))
print("Original image:\n",image)

# increase brightness
bright_image = image + 50
bright_image = np.clip(bright_image,0,255)

print("\nBright image:\n",bright_image)

# increase contrast
contrast_image = 1.5 * image
contrast_image = np.clip(contrast_image,0,255)

print("\nContrast image:\n",contrast_image.astype(int))
```
