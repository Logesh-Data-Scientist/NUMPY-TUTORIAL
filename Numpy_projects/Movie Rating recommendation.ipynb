MOVIE RATING RECOMMENDATION (Numpy Correlation Project)

~ Find movies with similar ratings using correlation

```python
import numpy as np
```

```python
np.random.seed(0)
ratings = np.random.randint(1,6,(5,4))

print("Movie ratings: ")
print(ratings)
```

```python
movie_corr = np.corrcoef(ratings.T)

print("Movie correlation matrix:\n",movie_corr,)
print()

similarities = movie_corr[0]

for i,score in enumerate(similarities):
    print(f"Movie{i + 1}:{score:.2f}")
```
