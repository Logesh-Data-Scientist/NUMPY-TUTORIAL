COMPANY SALES ANALYSIS

```python
import numpy as np 

months = np.array(["jan","feb","Mar","apr","may","june","jul","aug","sep","oct","nov","dec"])
sales = []

print("Enter the sales (in $1000) for each month: ")

for month in months :
    value = float(input(f"{month}"))
    sales.append(value)
    print(f"{month}\t:{value}$")

sales = np.array(sales)
print("------------------------------")
print("----COMPANY-SALES-ANALYSIS----")
print("------------------------------")
print()
print("total sales of the year\t:",np.sum(sales),"$")

mean = np.mean(sales)
print("average monthly sales\t:",round(mean,2),"$")
print("highest Sales\t\t:",np.max(sales),"$")
print("lowest sales\t\t:",np.min(sales),"$")

best_month = months[np.argmax(sales)]
print("best_month\t\t:",best_month)

worst_month = months[np.argmin(sales)]
print("worst_month\t\t:",worst_month)

above_average = months[sales > np.mean(sales)]
print("above_average\t\t:",above_average)

below_average = months[sales < np.mean(sales)]
print("below_average\t\t:",below_average)
```
