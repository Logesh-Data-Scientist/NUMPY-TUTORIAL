STUDENT MARK ANALYSIS

1. Average by subject and student
2. Highest & lowest marks per subject
3. Overall class topper
4. Pass count per subject
5. Which subject is difficult?
6. Ranking Students


students = rows

```python
import numpy as np

np.random.seed(25)
marks = np.random.randint(1,101 , size=(20,5))
print(marks)
print()

# 1. Average by subject and student

average_by_subjects = np.mean(marks , axis = 0)
print("average by subjects: ",average_by_subjects)
print()

average_by_students = np.mean(marks , axis = 1)
print("average by students: ",average_by_students)
print()

# 2. Highest & lowest per subject

Highest_value = np.max(marks,axis = 0)
lowest_value = np.min(marks,axis = 0)

print("highest marks of subjects: ",Highest_value)
print("lowest marks of subject: ",lowest_value)
print()

# 3. Overall class topper

total_marks = np.sum(marks, axis= 1)
topper = np.argmax(total_marks)

print("total marks :" ,total_marks)
print("topper of the class:",topper,"with marks of",total_marks[topper])
print()

# 4. Pass count per subject

pass_fail = marks >=35
pass_count = np.sum(pass_fail,axis = 0)

print("Pass count by subjects: ",pass_count)

# 5. Which subject is difficult?

diffcult_subject = np.argmin(average_by_subjects)
print("diffcult subject is",diffcult_subject,"with average marks of",average_by_subjects[diffcult_subject])  # index value

# 6. Ranking Students

ranks = np.argsort(-total_marks)
print("First Rank of class",ranks[0])
print("First Rank of class",ranks[:3])

```
