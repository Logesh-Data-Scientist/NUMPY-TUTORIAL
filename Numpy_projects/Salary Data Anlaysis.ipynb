SALARY DATA ANALYSIS USING NUMPY

generate random salary data and compute stats

```python
import numpy as np
```

```python
np.random.seed(0)
salaries = np.random.randint(20000,100000,size=1000)
```

```python
# print("Salaries: ", salaries)

print("Average salary\t\t\t: ", np.mean(salaries))
print("Median salary\t\t\t: ", np.median(salaries))
print("Salary standard deviation\t: ", np.std(salaries))
print("highest salary\t\t\t: ", np.max(salaries))
print("Lowest salary\t\t\t: ", np.min(salaries))   
```
